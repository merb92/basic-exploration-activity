```python
df.neighbourhood_group.unique()
```

```python
#Creating a smaller df to graph.
df_neighbourhood_group=df.groupby('neighbourhood_group').agg(['sum']).reset_index()
df_neighbourhood_group.plot(x='neighbourhood_group', y='price',kind="bar")
plt.xlabel('Neighbourhood_Group')
plt.ylabel('Total Earned From Air BNB Rental')
plt.title('Amount Earned in different neighbourhood group')
plt.show()
```
