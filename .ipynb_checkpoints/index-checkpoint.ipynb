```python
df = pd.read_csv('AB_NYC_2019.csv')
```

This dataset describes the listing activity and metrics in NYC, NY for 2019.

EDA and Visualization for Andy 

```python
df.neighbourhood_group.unique()
```

```python
#Creating a smaller df to graph.
df_neighbourhood_group=df.groupby('neighbourhood_group').agg(['sum']).reset_index()
df_neighbourhood_group.plot(x='neighbourhood_group', y='price',kind="bar")
plt.xlabel('Neighbourhood_Group')
plt.ylabel('Total Earned From Air BNB Rental')
plt.title('Amount Earned in different neighbourhood group')
plt.show()
```

EDA and Visualization for Mike

```python
brooklyn_df = df[df['neighbourhood_group'] == 'Brooklyn']
manhattan_df = df[df['neighbourhood_group'] == 'Manhattan']
queens_df = df[df['neighbourhood_group'] == 'Queens']
staten_island_df = df[df['neighbourhood_group'] == 'Staten Island']
bronx_df = df[df['neighbourhood_group'] == 'Bronx']
mike_fig = plt.figure(figsize=(10,10))
ax1 = mike_fig.add_subplot(3,2,1)
ax2 = mike_fig.add_subplot(3,2,2)
ax3 = mike_fig.add_subplot(3,2,3)
ax4 = mike_fig.add_subplot(3,2,4)
ax5 = mike_fig.add_subplot(3,2,5)
ax1.hist(brooklyn_df[brooklyn_df['minimum_nights'] < 30]['minimum_nights'])
ax1.set_title('Minimum Nights in Brooklyn Under 30 Nights')
ax2.hist(manhattan_df[manhattan_df['minimum_nights'] < 30]['minimum_nights'])
ax2.set_title('Minimum Nights in Manhattan Under 30 Nights')
ax3.hist(queens_df[queens_df['minimum_nights'] < 30]['minimum_nights'])
ax3.set_title('Minimum Nights in Queens Under 30 Nights')
ax4.hist(staten_island_df[staten_island_df['minimum_nights'] < 30]['minimum_nights'])
ax4.set_title('Minimum Nights in Staten Island Under 30 Nights')
ax5.hist(bronx_df[bronx_df['minimum_nights'] < 30]['minimum_nights'])
ax5.set_title('Minimum Nights in Bronx Under 30 Nights')
plt.show()
```

EDA and Visualization for Nadine

```python
import seaborn as sns
sns.catplot(x = 'room_type', kind = 'count',  hue = 'neighbourhood_group', palette = 'colorblind', data = df)
plt.title("Room types")
plt.xlabel("Room Types")
plt.ylabel("Number of listings")
plt.show()
```

```python
# In Manhattan entire home/appartment is the most common type of listing unlike other neighbourhood groups,
# where private room is most common.
# There are very few shared room listings, indicating AirBnB is not the best way to find such accomodation.
```
